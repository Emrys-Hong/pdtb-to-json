```python
from stanfordcorenlp import StanfordCoreNLP
import stanfordnlp
import re
import json
import spacy,en_core_web_sm
```

```python
parse = StanfordCoreNLP(r'/home/pengfei/documents/stanford-corenlp-full-2018-10-05/')
nlp = stanfordnlp.Pipeline()
tokenizor = en_core_web_sm.load()
```

```python
def constituent_parsing(sent):
    return parse.parse(sent).replace('ROOT', " ").replace('\n', '').replace(' ', '')

"""
@input relations: (<Word index=0;text=ROOT>,
 'root',
 <Word index=4;text=Inc.;lemma=Inc.;upos=PROPN;xpos=NNP;feats=Number=Sing;governor=0;dependency_relation=root>)

@return : ["root", "ROOT-0", "cut-16"]
"""
def relation_to_list(relation):
    return [relation[1], f'{relation[0].text}-{relation[0].index}', f'{relation[0].text}-{relation[0].index}']

def relations_to_list(relations):
    return [relation_to_list(r) for r in relations]

"""
@input filename: filename of raw text
@return name: the name of filename
@return jsonObject: the parsed json object"""
def pdtb_to_json(folder, name):
    with open(folder + name) as f:
        file_content = f.read()
        
        ## for each sentence in that document
        doc = tokenizor(file_content)
        sentences = []
        for idno, sent in enumerate(doc.sents):
            if idno == 0: continue ## skip first __START__ sentence
            sent = str(sent)
            sent_nlp = nlp(sent).sentences[0]
            relations = sent_nlp.dependencies
            dependencies = relations_to_list(relations)
            parsetree = constituent_parsing(sent)
            
            # for each word in sent
            words = []
            sent_index = file_content.index(sent)
            for word, token in zip(sent_nlp.words, tokenizor(sent)):
                begin = sent_index+token.idx
                token_dict = {'CharacterOffsetBegin':begin, 
                              "CharacterOffsetEnd": begin+len(token), 
                              "PartOfSpeech": word.pos}
                words.append([token.text, token_dict])
                
            sentences.append({"dependencies": dependencies, "parsetree": parsetree, "words": words})
            
            
    return name, {"sentences": sentences}
```

```python
if __name__ == "__main__":
    name, dictFile = pdtb_to_json('/home/pengfei/data/trial/raw_train/', 'wsj_1000')
    parse.close()
```
